**Data Processing**

```python
x = df[["Age","EstimatedSalary"]]
y = df.iloc[:,-1]
```

**Splitting**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=123)
```

**Standardization**

```python

sc=StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Logistic Regression

**Model preparation**

```python
logreg = LogisticRegression(solver='liblinear').fit(x_train,y_train)
y_pred = logreg.predict(x_test)
y_pred
```

```python
prob = logreg.predict_proba(x_test)
prob
```

**Evaluation**

```python
logreg.score(x_test,y_test)
```

```python
logreg.score(x_train,y_train)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred,labels = logreg.classes_)
print(cm)

tp, fn, fp, tn = confusion_matrix(y_test,y_pred,labels=[1,0]).reshape(-1)
```

```python
plt.title('Confusion matrix of Logistic Regression')
sns.heatmap(cm,annot=True,cmap='Blues',fmt='g')
plt.show()
```

```python
print('\nAccuracy: {:.2f}'.format(accuracy_score(y_test,y_pred)))
print('Error Rate: {:.2f}'.format((fp+fn)/(tp+tn+fn+fp)))
print('Sensitivity (Recall or True positive rate) :{:.3f}'.format(tp/(tp+fn)))
print('Specificity:{:.3f}'.format(tn/(fp+tn)))
print('Precision  :{:.2f}'.format(tp/(tp+fp)))
print('False Positive Rate: {:.3f}'.format(fp/(tn+fp)))
```

```python
from sklearn.model_selection import cross_val_score
accuracy_rate = []
for i in range(1,50):
    k = KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(k,x,y,cv=10)
    accuracy_rate.append(score.mean())
```

```python
print(accuracy_rate)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,50),accuracy_rate,color='Blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.ylabel('Accuracy rate')
plt.xlabel('K')
plt.title('Accuracy rate Vs. K')
```

```python
knn = KNeighborsClassifier(n_neighbors =10)
knn.fit(x_train,y_train)
y_pred2 = knn.predict(x_test)
```
